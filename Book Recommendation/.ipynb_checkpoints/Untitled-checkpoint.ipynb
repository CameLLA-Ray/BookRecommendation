```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 为图表设置样式
sns.set(style="whitegrid")

# 定义文件路径
users_file = 'Users.csv'
books_file = 'Books.csv'
ratings_file = 'Ratings.csv'

# 加载数据集
try:
    users = pd.read_csv(users_file)
    books = pd.read_csv(books_file)
    ratings = pd.read_csv(ratings_file)
except UnicodeDecodeError:
    print("UTF-8 编码失败，尝试使用 latin-1 编码...")
    users = pd.read_csv(users_file, encoding='latin-1')
    books = pd.read_csv(books_file, encoding='latin-1', on_bad_lines='skip') 
    ratings = pd.read_csv(ratings_file, encoding='latin-1')
except Exception as e:
    print(f"加载文件时发生错误: {e}")
    print("请确保 'Users.csv', 'Books.csv', 和 'Ratings.csv' 文件位于正确的目录中。")

print("数据集加载成功！")
```
